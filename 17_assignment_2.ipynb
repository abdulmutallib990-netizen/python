```python
# Q.1 Write a lambda function that takes a single number and returns its square.
     #Assign the lambda to a variable named square and call it with a few different values to verify it works.

sq= lambda x: x**2

print(sq(14))
print(sq(18))
```

```python
# Q.2 Given a list of (name, age) tuples, sort them in ascending order by age using sorted() with a lambda as the key argument.


people = [("Alice", 30), ("Bob", 25), ("Charlie", 35), ("Diana", 28)]

ascending = sorted(people, key = lambda x: x[1])


print('Ascending', ascending)
```

```python
# Q.3 sort the above list in descending order by age.

desending = sorted(people, key = lambda x: x[1], reverse = True)

print('Descending',desending)
```

```python
# Q.4 Write a lambda that takes a single integer and returns the string "even" if the number is divisible by 2, or "odd" otherwise.
#     Assign it to a variable named parity and test it on several values.

t = lambda num: 'Even' if num%2==0 else 'Odd'

print(t(58))
print(t(87))
print(t(103))
```

```python
# Q.5 Store four lambda functions in a dictionary under the keys "add", "sub", "mul", and "div". Each lambda should take two numbers and perform the corresponding arithmetic operation.
#     Use the dictionary to build a simple calculator that looks up the operation by key and applies it to two operands.
ops = {
    "add": lambda a, b: a + b,
    "sub": lambda a, b: a - b,
    "mul": lambda a, b: a * b,
    "div": lambda a, b: a / b,
}

a, b = 10, 3

for name, func in ops.items():
    print(f"{name}: {func(a, b)}")
```

```python
# Q.6 : Given a list of employee dictionaries, each with a "name" and a "salary" key, sort the list by salary in descending order.
#       Where two employees share the same salary, sort those entries by name in ascending alphabetical order.
#       Use a single lambda as the key argument.

employees = [
    {"name": "Alice", "salary": 70000},
    {"name": "Bob", "salary": 90000},
    {"name": "Charlie", "salary": 70000},
    {"name": "Diana", "salary": 90000}
]
sorted =sorted(employees, key= lambda e: (-e["salary"], e["name"]))

for employee in sorted:
    print(employee)
```

```python
# Q.7 Write a general multiply(x, n) function that returns x * n.
# Use functools.partial() to create two specialised functions from it: double(x), which always multiplies by 2, and triple(x), which always multiplies by 3. Call both with several values.

from functools import partial

def multiply(a,b):
  return a * b
double = partial(multiply, b=2)
triple = partial(multiply, b=3)

print(f"double(5) = {double(5)}")
print(f"double(9) = {double(9)}")
print(f"triple(4) = {triple(4)}")
print(f"triple(7) = {triple(7)}")
```

```python
# Q.8  Using map() with User-Defined Function
#      Apply a normal function to every element of a list using map().

def add_gst(price):
  return price + (price * .18)

prices = [100,200,300]

final_prices = list(map(add_gst, prices))

print(final_prices)
```

```python
# Q.9 Using reduce() with Initial Value.
#     Use reduce() with an initial value to add numbers.

from functools import reduce
numbers = [10, 20, 30]


total = reduce(lambda x, y: x + y, numbers, 100)
print(total)
```

```python
# Q.10 Using reduce() to Find Maximum Number
#      Find the largest number from a list using reduce().

numbers = [45, 12, 89, 34, 67]
maximum = reduce(lambda x, y: x if x > y else y, numbers)
print(maximum)
```
