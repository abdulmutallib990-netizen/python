```python
# 3. Transfer statement
# 1. break- exit from current loop
# 2. continue- skip current condition and continue the loop
# 3. pass - used to create an empty blocks
```

```python
f = ['mango','apple','cherry','apple','grapes','mango','banana']
for item in f:
  if item == 'grapes':
    break
  else:
    print('Process',item)
```

```python
# continue : will help to continue the loop and skip current condition or item

products = ['mobile','laptop','headphone','mouse','keyboard','mobile','laptop','headphone']

search_item = input('Enter item to search:')

for item in products:
  print('Checking:',item)

  if item == search_item:
    print('Product found:',item)
    continue
```

```python
candidate = [
    {'name':'gayatri','status':'present'},
    {'name':'ajay','status':'absent'},
    {'name':'gorakh','status':'present'},
    {'name':'tabassum','status':'absent'},
    {'name':'ajinkya','status':'present'},
    ]
  # give an assignment to the present candidates and skip absent candidates

for student in candidate:
  if student['status'] == 'absent':
    print(student['name'],'is absent')
    continue
  else:
    print('Give an assignment to:',student['name'])
```

```python
marks = [85,90,76,0,68,-1,88,96]

for mark in marks:

  if mark == -1:
    print('Invalid marks found, Stop checking')
    break

  if mark == 0:
    print('Absent Student, Skip this student')
    continue

  if mark < 40:
    pass

  print('Valid marks:,',mark)
```

# **List Comprehension**

*   **to create a new list in simple oneliner solution**
*   it uses the terneary opeartor to check if else option





```python
# syntax: [operation for var i sequence/iteration]
```

```python
d = [2,3,4,5,6,7]
# create a new list with squared numbers
[num*num for num in d]
```

```python
nms = ['Prasad','imran','gayatri','jay','bhavana']
[nm.title()for nm in nms]
```

```python
[(nm.title(),len(nm))for nm in nms]  # (name,len(name))
```

```python
 # conditional list comprehension
 # fetch those names whose len is >6
 #syntax:[oepration for var in sequnce/iteration if<condition>]
 # first element fetched in nm---> check condition len(nm)>6--.perform operation
 [nm.title()for nm in nms if len(nm)>6]
```
