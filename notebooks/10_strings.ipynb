# **Types of Arguments**

```python
# Positioal argument
# These are args. in which sequence order matters

def test(x,y,z):
  print(x,y,z)

test(1000,3000,2000)
test(2000,3000,1000)
```

```python
def info(name,age,sal):
  print('Your name is',name,'Your age is:',age,'& Your salary is:',sal)

info("viki,",25,50000)   # positional args
```

```python
info(60000,'prasad',24)    # when user dosnt follow order then output is not as per the expection
```

```python
# To solve above problem we have keyword args.
# which doesnt require to follow an order
# bcz it directly refers to key=value format

info(sal=60000,age=34,name='viki')
```

```python
info('Vikki',22)  # we must supply 3 values to 3 args
```

```python
info(name='Imran')     # in keyword also we need to supply the values = to number of args
```

```python
info('viki',33,None)
```

```python
# To solve above problem we have default aruments
# using default args we set the static values in declaration
# so when use is not supplying the values from calling then system will take these default value

def info(name='Guest',age=25,sal=35000):
  print('Your name is',name,'Your age is:',age,'& Your salary is:',sal)

info()
```

```python
info('prasad')
```

```python
# when user is sending the new vlaues then those will be taken
# else it will take default values
```

```python
info(25,'prasad')
```

```python
info(age=35,name='parasad')
```

```python
# Rule- We can use positional and keywords args but we should mention positonal args first n then
# go with keyword

info('vandana',sal=60000,age=22)
```

```python
# 4.Variable length args
# 1. Variable length positioanl args
# 2. Variable length keyword args
```

```python
# 1. Variable length positional args--> *args
# when use r doesnt have a strict requirement of sending the values to the arguments
# and values varies from one calling to another then use var. len. positonal args,
# we need to prefix the argument with *
def test(*n):
  print(n)

test(10)
```

```python
test(10,20,30)
test('python',35,456485764856,'c++')
```

```python
test()
```

```python
# 2.Varialble length keyword ars--> ** kwargs

def test (**x):
  print(x)

test(num1=10,num2=20,num3=30)
```

```python
test(name='Anmol',age=33,sal=47598)
```

```python
# when we want to combine both we can do so
# but remember the rule
# first we need positional n then keyword

def sample(*args,**kwrgs):
  print(args,kwrgs)


sample(10,20,30,num1=200,num2=600)
```

```python
a = [10,'',20,30,'','',40,50,'']
for i in a:
  if type(i) == str:
    continue
  else:
    print(i)
```

```python
for i in a:
  if type(i) != str:
    print(i)
```
