# **Lambda Function**

```python
# Annoymous / nameless function
# it is a one liner function
# syntax : lambda parameter /s:expression
```

```python
# squaring of numbers

sq = lambda num:num*num

sq(7)
```

```python
# only one expression is allowed in lambda
sq= lambda num:num*num, num+100    # here we are adding one more expression which is not allowed

sq(8)
```

```python
# lambda fuction by default returns the value of the expression
change = lambda name:name.upper()

result =change('prasad')
print(result)
```

# **Higher order function**

```python
# Higher order function- It is a fucntion who need another function as an input
# Function(function)
```

```python
# map, filter and reuce are higher order funcions
```

```python
# iterable- collection of elements--> str,list,tuple,set,dict,rane,array
# map(function,iterable)
k = [20,3,3,50,10]

# apply an operation over each element from an iterable

map (lambda num:num*num,k)
```

```python
0x7cc2fd3c40a0
```

```python
# Rule- when we are getting an output in the form of object then we have 2 choices
# choicel- typecast the object to see actual values

tuple(map(lambda num:num*num,k))
```

```python
# Choice 2- use for loop
out = map(lambda num:num*num,k)
for val in out:
  print(val)
```

```python
# can we use more args in lambda:YES

test = lambda a,b,c:a+b
test(234,454,45)
```

```python
test = lambda a,b,c:max(a,b,c)
test(234,454,45)
```

```python
# filter(function, iterable)
# this is condition fuction
# the value for which condition is true, will get fetched and operation will be performed if required
p = (12,-33,40,-8,99,-12)
# filter -ve numbers

list(filter(lambda num:num>0,p))  # +ve numbers
```
