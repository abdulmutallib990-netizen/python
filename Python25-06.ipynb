# **Function with return keyword**

```python
# funtion will work on specific requirment
# it will generate one result
# this result we need to use as an input to the outside of the function
# for this purpose we need to use return keyword

def test(balance):
  if balance <=0:
    return 'No'
  else:
    return'Yes'

result = test(120)

if result=='Yes':
  print('You can perform Transacion')
else:
  print('You can\'t perform transaction')
# hence . when we want to bring the result of a function from inside to outside then
# we need return to be used
```

```python
from typing import AnyStr
# Purchase and billing example using return
items = {'rice':56,'chana':30,'wheat':60,'parle_g':10,'milk':45}
bill = []
def purchase(*args):
  for item in args:
    #print(item,item[item])
    bill.append(items[item])
  return bill

def check(total):
  return sum(total)

# calling of purchase
total= purchase('wheat','parle_g','milk')
print('Item amount:',total)

#calling of check
ans = check(total)
print('Final bill amount is Rs:',ans)
```

```python
# calculate % and calculate grade

def calculate_per(marks_obtaiened,total_marks):
  return(marks_obtaiened/total_marks)*100
#print (calculate_per(450,600))

def calculate_grade(per):
  if per >=75:
    return 'Distinction'
  elif per >= 65:
    return 'First class'
  elif per>= 55:
    return 'Second class'
  elif per>=45:
    return 'Pass'
  else:
    return 'Fail'

per = calculate_per(260,500)
result = calculate_grade(per)
print('Total percntage is:',per)
print('Your grade is:',result)
```
